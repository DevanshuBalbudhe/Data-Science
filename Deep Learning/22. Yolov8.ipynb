```python
from ultralytics import YOLO
import cv2
from PIL import Image
```

```python
print("\n" + "="*60)
print("TASK 1: OBJECT DETECTION")
print("="*60)
```

```python
# Load pre-trained YOLOv8 small model for object detection
model_detect = YOLO('yolov8s.pt')
```

```python
# Print model parameters
print(f"\nModel Type: Object Detection")
print(f"Model Name: {model_detect.model_name}")
print(f"Total Parameters: {sum(p.numel() for p in model_detect.model.parameters()):,}")
print(f"Trainable Parameters: {sum(p.numel() for p in model_detect.model.parameters() if p.requires_grad):,}")
```

```python
# Predict on new image
results_detect = model_detect.predict(
    source='/content/traffic.jpg',  # Replace with your image path
    save=True,
    conf=0.5,  # Confidence threshold
    iou=0.45   # IoU threshold for NMS
)
```

```python
# Print prediction results
print("\n--- DETECTION RESULTS ---")
for result in results_detect:
    boxes = result.boxes
    print(f"Total Objects Detected: {len(boxes)}")
    for idx, box in enumerate(boxes):
        class_id = int(box.cls[0])
        class_name = result.names[class_id]
        confidence = float(box.conf[0])
        x1, y1, x2, y2 = box.xyxy[0].tolist()
        print(f"\nObject {idx+1}:")
        print(f"  Class: {class_name}")
        print(f"  Confidence: {confidence:.4f}")
        print(f"  Bounding Box: ({x1:.2f}, {y1:.2f}, {x2:.2f}, {y2:.2f})")
```

```python
print("\n" + "="*60)
print("TASK 2: IMAGE CLASSIFICATION")
print("="*60)
```

```python
# Load pre-trained YOLOv8 small model for classification
model_cls = YOLO('yolov8s-cls.pt')
```

```python
# Print model parameters
print(f"\nModel Type: Image Classification")
print(f"Model Name: {model_cls.model_name}")
print(f"Total Parameters: {sum(p.numel() for p in model_cls.model.parameters()):,}")
print(f"Trainable Parameters: {sum(p.numel() for p in model_cls.model.parameters() if p.requires_grad):,}")
```

```python
# Predict on new image
results_cls = model_cls.predict(
    source='/content/agri.jpg',  # Replace with your image path
    save=True
)
```

```python
# Print prediction results
print("\n--- CLASSIFICATION RESULTS ---")
for result in results_cls:
    probs = result.probs
    top5_indices = probs.top5
    top5_conf = probs.top5conf.tolist()
    print(f"Top 5 Predictions:")
    for idx, (class_idx, conf) in enumerate(zip(top5_indices, top5_conf)):
        class_name = result.names[class_idx]
        print(f"  {idx+1}. {class_name}: {conf:.4f}")
```

```python
print("\n" + "="*60)
print("TASK 3: INSTANCE SEGMENTATION")
print("="*60)
```

```python
# Load pre-trained YOLOv8 small model for segmentation
model_seg = YOLO('yolov8s-seg.pt')
```

```python
# Print model parameters
print(f"\nModel Type: Instance Segmentation")
print(f"Model Name: {model_seg.model_name}")
print(f"Total Parameters: {sum(p.numel() for p in model_seg.model.parameters()):,}")
print(f"Trainable Parameters: {sum(p.numel() for p in model_seg.model.parameters() if p.requires_grad):,}")
```

```python
# Predict on new image
results_seg = model_seg.predict(
    source='/content/market.jpg',  # Replace with your image path
    save=True,
    conf=0.5,
    iou=0.45
)
```

```python
# Print prediction results
print("\n--- SEGMENTATION RESULTS ---")
for result in results_seg:
    if result.masks is not None:
        masks = result.masks
        boxes = result.boxes
        print(f"Total Objects Segmented: {len(masks)}")
        for idx, (mask, box) in enumerate(zip(masks, boxes)):
            class_id = int(box.cls[0])
            class_name = result.names[class_id]
            confidence = float(box.conf[0])
            print(f"\nObject {idx+1}:")
            print(f"  Class: {class_name}")
            print(f"  Confidence: {confidence:.4f}")
            print(f"  Mask Shape: {mask.data.shape}")
    else:
        print("No objects segmented in the image")
```

```python
print("\n" + "="*60)
print("TASK 4: POSE ESTIMATION")
print("="*60)
```

```python
# Load pre-trained YOLOv8 small model for pose estimation
model_pose = YOLO('yolov8s-pose.pt')
```

```python
# Print model parameters
print(f"\nModel Type: Pose Estimation")
print(f"Model Name: {model_pose.model_name}")
print(f"Total Parameters: {sum(p.numel() for p in model_pose.model.parameters()):,}")
print(f"Trainable Parameters: {sum(p.numel() for p in model_pose.model.parameters() if p.requires_grad):,}")
```

```python
# Predict on new image
results_pose = model_pose.predict(
    source='/content/avengers.webp',  # Replace with your image path
    save=True,
    conf=0.5
)
```

```python
# Print prediction results
print("\n--- POSE ESTIMATION RESULTS ---")
for result in results_pose:
    if result.keypoints is not None:
        keypoints = result.keypoints
        boxes = result.boxes
        print(f"Total Persons Detected: {len(boxes)}")
        for idx, (kpts, box) in enumerate(zip(keypoints, boxes)):
            confidence = float(box.conf[0])
            print(f"\nPerson {idx+1}:")
            print(f"  Confidence: {confidence:.4f}")
            print(f"  Number of Keypoints: {kpts.data.shape[1]}")
            print(f"  Keypoints Shape: {kpts.data.shape}")
    else:
        print("No persons detected in the image")
```
