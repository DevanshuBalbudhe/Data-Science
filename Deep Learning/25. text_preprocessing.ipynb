```python
1. Lowercasing (e.g., "Hello" → "hello")

Explanation: Converts all characters to lowercase to ensure consistency. Words like "Hello", "hello", and "HELLO" are treated as the same token, reducing vocabulary size and preventing models from seeing them as distinct. This is crucial for case-insensitive tasks like sentiment analysis but may be skipped for proper nouns (e.g., names) if context matters.

When to Use: Always, unless the task requires case sensitivity (e.g., NER for named entities).
Implementation: Use str.lower() in Python.

Pros: Simple, reduces duplicates. Cons: Loses information like acronyms (e.g., "US" vs. "us").
```

```python
2. Removing Extra Spaces & Line Breaks

Explanation: Raw text often has multiple spaces, tabs, or newlines from formatting. This step replaces them with single spaces and strips leading/trailing whitespaces to make text compact and uniform.

When to Use: Essential for all text to avoid tokenization errors (e.g., "word  " becoming two tokens).
Implementation: Use re.sub(r'\s+', ' ', text).strip() to replace multiple whitespaces (including \n, \t) with a single space.

Pros: Improves readability and efficiency. Cons: May merge intended separations in poetry or code
```

```python
3. Removing Punctuation (e.g., ".,!?")

Explanation: Punctuation like periods, commas, or exclamation marks often doesn't add semantic value in NLP tasks. Removing them simplifies text but can be optional if punctuation aids tasks like sentence boundary detection.

When to Use: For bag-of-words models; keep for sequence models like RNNs if needed.
Implementation: Use re.sub(r'[^\w\s]', '', text) or string.punctuation to remove non-alphanumeric characters except spaces.

Pros: Reduces noise. Cons: Loses emphasis (e.g., "!!" for excitement).
```

```python
4. Removing Numbers (Optional)

Explanation: Numbers (e.g., "123") may not be relevant in text analysis unless the task involves quantities (e.g., financial NLP). This is optional—replace with placeholders like "<NUM>" if needed.

When to Use: When numbers are noise (e.g., timestamps); keep for tasks like date extraction.
Implementation: Use re.sub(r'\d+', '', text) to remove digits.

Pros: Simplifies vocabulary. Cons: Loses quantitative info.
```

```python
5. Removing Special Characters (e.g., "@#$%^")

Explanation: Characters like @, #, $, % are common in usernames, hashtags, or currencies but often irrelevant. This overlaps with punctuation removal but targets non-standard symbols.

When to Use: For social media text; keep hashtags if analyzing trends.
Implementation: Use re.sub(r'[^a-zA-Z0-9\s]', '', text) to keep only letters, numbers, and spaces.

Pros: Cleans noisy data. Cons: May remove useful symbols in domain-specific text.
```

```python
6. Removing HTML Tags (e.g., " ")

Explanation: Web-scraped text contains tags like ,
,  which are markup, not content. Removing them extracts pure text.

When to Use: For data from websites or emails.
Implementation: Use re.sub(r'<.*?>', '', text) to strip tags.

Pros: Reveals actual content. Cons: May leave artifacts if tags are malformed.
```

```python
7. Removing URLs & Emails

Explanation: Links (e.g., "https://example.com") and emails (e.g., "user@example.com") are often placeholders or spam, not meaningful for analysis.

When to Use: Always for general cleaning; replace with "<URL>" or "<EMAIL>" if tracking.
Implementation: Use regex like re.sub(r'http\S+|www\S+|mailto:\S+|\S+@\S+', '', text).

Pros: Reduces irrelevant tokens. Cons: Loses reference info.
```

```python
8. Removing Emojis & Emoticons

Explanation: Emojis (😊) and emoticons (:)) add sentiment but are non-textual. Remove for text-only models; encode or keep for multimodal tasks.

When to Use: For traditional NLP; analyze separately for sentiment.
Implementation: Use re.sub(r'[^\x00-\x7F]+', '', text) for basic removal, or libraries like emoji for precise handling.

Pros: Standardizes to ASCII. Cons: Loses emotional cues.
```

```python
9. Handling Contractions (e.g., "don't" → "do not")

Explanation: Expands shortened forms (e.g., "I'm" → "I am") for better tokenization and understanding. This helps models treat them as full words.

When to Use: For English text; essential for consistency.
Implementation: Use the contractions library: contractions.fix(text).

Pros: Improves semantics. Cons: May not handle slang well.
```

```python
10. Removing Non-ASCII Characters

Explanation: Non-ASCII (e.g., accented letters like é, or foreign scripts) can cause encoding issues. Remove or normalize to ASCII for English-focused NLP.

When to Use: For monolingual English data; use Unicode normalization otherwise.
Implementation: Use text.encode('ascii', 'ignore').decode('ascii') or re.sub(r'[^\x00-\x7F]+', '', text).

Pros: Avoids encoding errors. Cons: Loses multilingual info.
```

```python
import re
import string
import contractions
```

```python
def basic_text_cleaning(text):
    # Handle contractions
    text = contractions.fix(text)
    # Lowercasing
    text = text.lower()
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove URLs and emails
    text = re.sub(r'http\S+|www\S+|mailto:\S+|\S+@\S+', '', text)
    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', text)
    # Remove numbers (optional)
    text = re.sub(r'\d+', '', text)
    # Remove emojis/non-ASCII
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    # Remove extra spaces & line breaks
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
### Tokenization
```

```python
1. Sentence Tokenization
How would you split a paragraph into individual sentences?

What clues (punctuation, capitalization, etc.) tell us where one sentence ends and another begins?

Purpose: Useful when we want to process text one sentence at a time (e.g., machine translation, summarization).

2. Word Tokenization

Once we have a sentence, how do we break it into words?

Is "don't" one word or two? What about "New York" or "ice-cream"?

Why might simple splitting on spaces fail?

Purpose: Most traditional NLP tasks (like bag-of-words, TF-IDF) rely on word tokens.

3. Subword Tokenization (The Modern/Advanced Approach)
Here's where it gets fascinating.
Large language models (like BERT, GPT) have a fixed vocabulary size (often 30k–50k tokens).

How can they handle rare words, misspellings, or words in other languages without "unknown" tokens everywhere?
This is where subword methods shine:

Byte-Pair Encoding (BPE): Starts with characters, repeatedly merges the most frequent pairs. Used in GPT models.

WordPiece: Similar to BPE but maximizes likelihood when adding new subwords. Used in BERT.

SentencePiece: Language-agnostic, treats text as raw characters (great for multilingual models).

Question: Why do you think subword tokenization handles out-of-vocabulary (OOV) words better than pure word tokenization?
Libraries We'll Use

Common tools:

nltk → simple and educational
spaCy → fast and accurate, especially for sentence splitting
transformers (from Hugging Face) → gives us industry-grade subword tokenizers (BERT, GPT-2, etc.)
```

```python
## Stemming
```

```python
First, pause and reflect: Why do we want to reduce "playing," "plays," "played" all to something like "play"? What problem in text analysis does this solve?
Think about search engines or topic modeling: Should "fish," "fishing," "fished," "fisher" be treated as different words or the same concept? Stemming aggressively groups them to reduce vocabulary size and improve matching.
But here's the trade-off: Stemming is rough—it chops words based on rules, so "studies" → "studi" (not "study"), "better" → "better" (wait, no—"bett"), or "news" → "new". Grammar gets sacrificed for speed and simplicity.
Key Questions to Ponder:

When might inaccurate stems hurt your application (e.g., chatbots vs. information retrieval)?
Is stemming still widely used in the era of BERT/GPT? Why or why not?

Types of Stemmers

Porter Stemmer (1980) – Most popular, balanced, gentle rules. Widely used.
Snowball Stemmer – Improved Porter, supports many languages, slightly more aggressive.
Lancaster Stemmer – Very aggressive, fastest, but often over-stems (e.g., "woman" → "wom").

Question: Which would you choose for English tweets (short, noisy text) vs. formal documents?
Libraries
Mainly nltk (all three available). spaCy doesn't include stemming by default (it prefers lemmatization—we'll cover that next!).
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

words = ["play", "playing", "played", "plays"]
for w in words:
    print(w, "→", ps.stem(w))
```

```python
from nltk.stem import SnowballStemmer
snow = SnowballStemmer("english")

words = ["running", "runs", "runner", "easily", "fairly"]
for w in words:
    print(w, "→", snow.stem(w))
```

```python
from nltk.stem import LancasterStemmer
ls = LancasterStemmer()

words = ["studies", "studying", "study", "student"]
for w in words:
    print(w, "→", ls.stem(w))
```

```python
### Lemmetization
```

```python
Why do you think many people call it "better than stemming"? What was the biggest frustration you noticed with stemming in the previous examples (like "studies" → "studi" or "better" → "bett")?
Lemmatization aims to reduce words to their canonical dictionary form (called the lemma)—a real, meaningful word—rather than a rough stem.
Examples:

running → run
better/best → good
geese → goose
ate → eat

The magic? It uses vocabulary + morphological rules and often Part-of-Speech (POS) tags to make smarter decisions.
Reflection Questions:

Why does knowing if "better" is an adjective matter for getting "good"?
Can lemmatization work accurately without POS tags? What happens if we guess wrong?
In what tasks (sentiment, chatbots, translation) would real words matter more than stems?

Key Advantages Over Stemming

More accurate and human-readable output
Handles irregular forms (go → go, went → go)
Context-aware (with POS)

Libraries

nltk → WordNet Lemmatizer (classic, needs POS)
spaCy → Built-in, fast, accurate, uses POS automatically
textblob, pattern (alternatives)
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')      # only once
nltk.download('omw-1.4')      # optional extended data

lemmatizer = WordNetLemmatizer()

words = ["running", "runs", "ran", "better", "best", "geese", "children"]
for w in words:
    print(w, "→", lemmatizer.lemmatize(w))
```

```python
  First, reflect: Why do real-world texts contain contractions ("can't"), typos ("recieve"), elongated words ("yesssss"), or accented characters ("naïve")? How might these variations confuse models that treat "can't" and "cannot" as completely different tokens?
Normalization helps reduce noise and vocabulary explosion by bringing equivalent expressions to the same form.
Common techniques we'll explore:

Expanding contractions (can't → cannot)
Spelling correction (recieve → receive)
Handling repeated/reduplicated characters (soooo → so, happpy → happy)
Removing/replacing accented characters (café → cafe, résumé → resume)

Thought-provoking questions:

When might expanding contractions hurt (e.g., sentiment: "can't wait!" vs "cannot wait!")?
Is perfect spelling correction always needed, or could it change meaning ("lead" metal vs "led")?
In informal text (tweets, chats), should we aggressively normalize elongations?
```

```python
contractions = {
    "can't": "cannot", "won't": "will not", "I'm": "I am", "don't": "do not",
    "it's": "it is", "you're": "you are", "he's": "he is", "she's": "she is",
    "we're": "we are", "they're": "they are", "isn't": "is not", "aren't": "are not"
}

text = "I can't believe it's happening. You're awesome!"
words = text.split()
expanded = [contractions.get(word, word) for word in words]
print("Original :", text)
print("Expanded :", " ".join(expanded))
```

```python
import unicodedata

def remove_accents(text):
    return ''.join(c for c in unicodedata.normalize('NFD', text)
                   if unicodedata.category(c) != 'Mn')

text = "café naïve résumé Pokémon"
normalized = remove_accents(text)
print("Original  :", text)
print("No accents:", normalized)
```

```python
### Noise Handling
```

```python
Think about where most text data comes from today—tweets, chats, comments, reviews, forums. Is that text clean and formal like a textbook? Or full of shortcuts, emotions, creativity, and errors?
That's noise: anything that deviates from standard, formal language.
Common types we'll explore:

Slang words (gr8 → great, cuz → because)
Chat abbreviations (lol, brb, btw, asap, omg)
Misspellings (recieve → receive, teh → the)
Emojis & emoticons (😊, :), XD)
Repeated punctuation (!!!, ???)
Code-mixed language (Hinglish: "I am bahut tired yaar", Spanglish, etc.)

Deep questions to reflect on:

Is all "noise" actually bad? Does "lol" or "gr8" carry emotional or stylistic meaning?
In sentiment analysis of tweets, should you convert "gr8" to "great" or keep it?
For Hinglish text ("This movie was mast!"), can English-only models understand it without help?
When does cleaning noise improve performance—and when does it remove valuable signals?
```

```python
slang_dict = {
    "gr8": "great", "g8": "great", "cuz": "because", "bcuz": "because",
    "u": "you", "ur": "your", "plz": "please", "thx": "thanks",
    "lol": "laughing out loud", "lmao": "laughing my ass off",
    "btw": "by the way", "asap": "as soon as possible", "brb": "be right back"
}

text = "gr8 job btw! lol ur awesome"
words = text.lower().split()
cleaned = [slang_dict.get(word, word) for word in words]
print("Original:", text)
print("Cleaned :", " ".join(cleaned))
```
