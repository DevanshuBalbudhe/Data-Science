# **Lenet-5**

```python
import tensorflow as tf
print(tf.config.list_physical_devices('GPU'))
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
x_train.shape
```

```python
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)
```

```python
x_train[0].shape
```

```python
model = tf.keras.Sequential()

# Layer 1: Conv + ReLU + MaxPool
model.add(tf.keras.layers.Conv2D(6, kernel_size=5, activation='relu', input_shape=(28, 28, 1)))
model.add(tf.keras.layers.MaxPooling2D(pool_size=2))

# Layer 2: Conv + ReLU + MaxPool
model.add(tf.keras.layers.Conv2D(16, kernel_size=5, activation='relu'))
model.add(tf.keras.layers.MaxPooling2D(pool_size=2))

# Flatten karo
model.add(tf.keras.layers.Flatten())

# Fully Connected Layers
model.add(tf.keras.layers.Dense(120, activation='relu'))
model.add(tf.keras.layers.Dense(84, activation='relu'))
model.add(tf.keras.layers.Dense(10, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train, y_train, epochs=10, batch_size=128, validation_data=(x_test, y_test), validation_split= 0.25)
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f'Test Accuracy: {test_acc}')
```

```python
model.save('lenet5_fashion_mnist.h5')
```

```python
model = tf.keras.models.load_model('lenet5_fashion_mnist.h5')

(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0

predictions = model.predict(x_test[:10])

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

print("Predictions:")
for i in range(10):
    predicted_class = np.argmax(predictions[i])
    actual_class = y_test[i]
    print(f"Image {i+1}: Predicted = {class_names[predicted_class]}, Actual = {class_names[actual_class]}")


y_test_categorical = tf.keras.utils.to_categorical(y_test, 10)
test_loss, test_accuracy = model.evaluate(x_test, y_test_categorical)
print(f"\nOverall Test Accuracy: {test_accuracy * 100:.2f}%")


single_image = x_test[100:101]
single_prediction = model.predict(single_image)
predicted_label = np.argmax(single_prediction)
print(f"\nSingle Image Prediction: {class_names[predicted_label]}")
```

# **VGG-19**

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
device
```

```python
transform_train = transforms.Compose([transforms.RandomHorizontalFlip(), transforms.RandomCrop(32, padding=4), transforms.ToTensor(), transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
```

```python
transform_test = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
```

```python
trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
```

```python
testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)
```

```python
trainloader = DataLoader(trainset, batch_size=64, shuffle=True)
```

```python
testloader = DataLoader(testset, batch_size=64, shuffle=False)
```

```python
model = nn.Sequential()

# Block 1 - Conv layer 1
model.add_module('conv1_1', nn.Conv2d(3, 64, kernel_size=3, padding=1))
model.add_module('relu1_1', nn.ReLU(inplace=True))

# Block 1 - Conv layer 2
model.add_module('conv1_2', nn.Conv2d(64, 64, kernel_size=3, padding=1))
model.add_module('relu1_2', nn.ReLU(inplace=True))

# Block 1 - Max pooling
model.add_module('pool1', nn.MaxPool2d(kernel_size=2, stride=2))

# Block 2 - Conv layer 1
model.add_module('conv2_1', nn.Conv2d(64, 128, kernel_size=3, padding=1))
model.add_module('relu2_1', nn.ReLU(inplace=True))

# Block 2 - Conv layer 2
model.add_module('conv2_2', nn.Conv2d(128, 128, kernel_size=3, padding=1))
model.add_module('relu2_2', nn.ReLU(inplace=True))

# Block 2 - Max pooling
model.add_module('pool2', nn.MaxPool2d(kernel_size=2, stride=2))

# Block 3 - Conv layer 1
model.add_module('conv3_1', nn.Conv2d(128, 256, kernel_size=3, padding=1))
model.add_module('relu3_1', nn.ReLU(inplace=True))

# Block 3 - Conv layer 2
model.add_module('conv3_2', nn.Conv2d(256, 256, kernel_size=3, padding=1))
model.add_module('relu3_2', nn.ReLU(inplace=True))

# Block 3 - Conv layer 3
model.add_module('conv3_3', nn.Conv2d(256, 256, kernel_size=3, padding=1))
model.add_module('relu3_3', nn.ReLU(inplace=True))

# Block 3 - Conv layer 4
model.add_module('conv3_4', nn.Conv2d(256, 256, kernel_size=3, padding=1))
model.add_module('relu3_4', nn.ReLU(inplace=True))

# Block 3 - Max pooling
model.add_module('pool3', nn.MaxPool2d(kernel_size=2, stride=2))

# Block 4 - Conv layer 1
model.add_module('conv4_1', nn.Conv2d(256, 512, kernel_size=3, padding=1))
model.add_module('relu4_1', nn.ReLU(inplace=True))

# Block 4 - Conv layer 2
model.add_module('conv4_2', nn.Conv2d(512, 512, kernel_size=3, padding=1))
model.add_module('relu4_2', nn.ReLU(inplace=True))

# Block 4 - Conv layer 3
model.add_module('conv4_3', nn.Conv2d(512, 512, kernel_size=3, padding=1))
model.add_module('relu4_3', nn.ReLU(inplace=True))

# Block 4 - Conv layer 4
model.add_module('conv4_4', nn.Conv2d(512, 512, kernel_size=3, padding=1))
model.add_module('relu4_4', nn.ReLU(inplace=True))

# Block 4 - Max pooling
model.add_module('pool4', nn.MaxPool2d(kernel_size=2, stride=2))

# Block 5 - Conv layer 1
model.add_module('conv5_1', nn.Conv2d(512, 512, kernel_size=3, padding=1))
model.add_module('relu5_1', nn.ReLU(inplace=True))

# Block 5 - Conv layer 2
model.add_module('conv5_2', nn.Conv2d(512, 512, kernel_size=3, padding=1))
model.add_module('relu5_2', nn.ReLU(inplace=True))

# Block 5 - Conv layer 3
model.add_module('conv5_3', nn.Conv2d(512, 512, kernel_size=3, padding=1))
model.add_module('relu5_3', nn.ReLU(inplace=True))

# Block 5 - Conv layer 4
model.add_module('conv5_4', nn.Conv2d(512, 512, kernel_size=3, padding=1))
model.add_module('relu5_4', nn.ReLU(inplace=True))

# Block 5 - Max pooling
model.add_module('pool5', nn.MaxPool2d(kernel_size=2, stride=2))

# Flatten layer
model.add_module('flatten', nn.Flatten())

# Fully connected layer 1
model.add_module('fc1', nn.Linear(512 * 1 * 1, 4096))
model.add_module('relu6', nn.ReLU(inplace=True))
model.add_module('dropout1', nn.Dropout(0.5))

# Fully connected layer 2
model.add_module('fc2', nn.Linear(4096, 4096))
model.add_module('relu7', nn.ReLU(inplace=True))
model.add_module('dropout2', nn.Dropout(0.5))

# Output layer - 10 classes for CIFAR-10
model.add_module('fc3', nn.Linear(4096, 10))

model = model.to(device)

criterion = nn.CrossEntropyLoss()

optimizer = optim.Adam(model.parameters(), lr=0.001)

data_iter = iter(trainloader)
inputs, labels = next(data_iter)
inputs, labels = inputs.to(device), labels.to(device)

model.train()

optimizer.zero_grad()

outputs = model(inputs)

loss = criterion(outputs, labels)

loss.backward()

optimizer.step()

print(f'First batch loss: {loss.item():.4f}')

torch.save(model.state_dict(), 'vgg19_cifar10.pth')
print('Model saved!')
```
