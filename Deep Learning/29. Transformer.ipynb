## **1. SENTIMENT ANALYSIS (Positive/Negative)**

```python
sentiment_analyzer = pipeline("sentiment-analysis")
```

```python

texts = [
    "I love this product!",
    "This is terrible and waste of money",
    "It's okay, nothing special"
]
```

```python
sentiment_analyzer( "I love this product!")[0]
```

```python
for text in texts:
    result = sentiment_analyzer(text)[0]
    print(f"Text: {text}")
    print(f"Result: {result['label']} (Score: {result['score']:.2f})\n")
```

## **3. QUESTION ANSWERING**

```python
qa_model = pipeline("question-answering")
```

```python
context = """
The Taj Mahal is an ivory-white marble mausoleum on the right bank of
the river Yamuna in the Indian city of Agra. It was commissioned in 1632
by the Mughal emperor Shah Jahan to house the tomb of his favourite wife,
Mumtaz Mahal.
"""
```

```python
questions = [
    "Where is Taj Mahal located?",
    "Who built the Taj Mahal?",
    "When was it commissioned?"
]
```

```python
for question in questions:
    result = qa_model(question=question, context=context)
    print(f"Q: {question}")
    print(f"A: {result['answer']} (Confidence: {result['score']:.2f})\n")
```

## **4. TEXT SUMMARIZATION (Fixed)**

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "facebook/bart-large-cnn"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
long_text = """
Climate change is one of the most pressing issues facing our planet today.
Rising global temperatures are causing ice caps to melt, sea levels to rise,
and weather patterns to become more extreme. Scientists worldwide agree that
human activities, particularly the burning of fossil fuels, are the primary
cause of recent climate change. The effects are already being felt across
the globe, from more frequent hurricanes to prolonged droughts. Immediate
action is needed to reduce greenhouse gas emissions and transition to
renewable energy sources.
"""
```

```python
inputs = tokenizer(long_text, return_tensors="pt",
                   max_length=1024, truncation=True)
```

```python
summary_ids = model.generate(
    inputs["input_ids"],
    max_length=100,
    min_length=25,
    length_penalty=2.0,
    num_beams=4,
    early_stopping=True
)
```

```python
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print(f"Original length: {len(long_text)} characters")
print(f"\nSummary: {summary}\n")
```

## **5. TRANSLATION (English to French)**

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
model_name = "Helsinki-NLP/opus-mt-en-fr"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
english_texts = [
    "Hello, how are you?",
    "I love learning AI",
    "This is amazing!"
]
```

```python
for text in english_texts:
    inputs = tokenizer(text, return_tensors="pt", padding=True)
    translated = model.generate(**inputs)
    result = tokenizer.decode(translated[0], skip_special_tokens=True)
    print(f"English: {text}")
    print(f"French: {result}\n")
```

## **6. NAMED ENTITY RECOGNITION (NER)**

```python
ner_model = pipeline("ner", aggregation_strategy="simple")
```

```python
text = "Apple Inc. was founded by Steve Jobs in Cupertino, California. Elon Musk runs Tesla and SpaceX."
entities = ner_model(text)
```

```python
print(f"Text: {text}\n")
print("Entities found:")
for entity in entities:
    print(f"  - {entity['word']}: {entity['entity_group']} (Score: {entity['score']:.2f})")
```

## **7. FILL MASK (Word Prediction)**

```python
fill_mask = pipeline("fill-mask", model="bert-base-uncased")
```

```python
sentence = "The capital of France is [MASK]."
predictions = fill_mask(sentence)
```

```python
print(f"Sentence: {sentence}\n")
print("Top predictions:")
for i, pred in enumerate(predictions[:6], 1):
    print(f"{i}. {pred['token_str']} (Score: {pred['score']:.2f})")
```

## **8. ZERO-SHOT CLASSIFICATION**

```python
zero_shot = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
text = "I just bought a new laptop with great performance"
labels = ["technology", "sports", "politics", "cooking"]
```

```python
result = zero_shot(text, candidate_labels=labels)
```

```python
print(f"Text: {text}\n")
print("Classification results:")
for label, score in zip(result['labels'], result['scores']):
    print(f"  - {label}: {score:.2%}")
```

## **10. ENGLISH TO HINDI TRANSLATION**

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
for sentence in english_sentences:
    inputs = tokenizer(sentence, return_tensors="pt", padding=True)
    translated = model.generate(**inputs)
    result = tokenizer.decode(translated[0], skip_special_tokens=True)
    print(f"English: {sentence}")
    print(f"Hindi: {result}\n")
```
